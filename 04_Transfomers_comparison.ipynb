```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Transformers Comparison
```

```python
bert_results = {
    "Model": [
        "DistilBERT",
        "RoBERTa",
        "BioBERT",
        "ClinicalBERT",
        "PubMedBERT",
        "MentalBERT"
    ],
    "Accuracy": [0.82, 0.84, 0.81, 0.83, 0.83, 0.84],   # <- update with your final scores
    "Macro F1": [0.81, 0.83, 0.80, 0.82, 0.82, 0.84]   # <- update with your final scores
}

bert_df = pd.DataFrame(bert_results)

# Sort by Accuracy for consistency
bert_df_acc = bert_df.sort_values("Accuracy", ascending=False)
bert_df_f1 = bert_df.sort_values("Macro F1", ascending=False)

# Accuracy Plot
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="Model", y="Accuracy", data=bert_df_acc, palette="viridis",hue='Model')

plt.title("Transformer Model Comparison - Accuracy", fontsize=14, weight='bold')
plt.xlabel("Models", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.ylim(0.78, 0.87)  # tighter range to highlight differences
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Add values on bars
for i, v in enumerate(bert_df_acc["Accuracy"]):
    plt.text(i, v + 0.003, f"{v:.3f}", ha='center', fontsize=11)

plt.tight_layout()
plt.show()

# Macro F1 Plot
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="Model", y="Macro F1", data=bert_df_f1, palette="viridis",hue='Model')

plt.title("Transformer Model Comparison - Macro F1 Score", fontsize=14, weight='bold')
plt.xlabel("Models", fontsize=12)
plt.ylabel("Macro F1 Score", fontsize=12)
plt.ylim(0.78, 0.87)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Add values on bars
for i, v in enumerate(bert_df_f1["Macro F1"]):
    plt.text(i, v + 0.003, f"{v:.3f}", ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```
